```python
df = sb.events(match_id=7584)
```

```python
df.columns
```

```python
df = df[['timestamp', 'period', 'minute','team', 'shot_outcome', 'shot_statsbomb_xg', 'type']]
```

```python
pd.set_option('display.max_rows', 10)
df
```

```python
df = df[df.type == 'Shot']
```

```python
df
```

```python
def create_xticks(period, minute):
    if period == 1:
        return '1st-' + str(minute)
    else:
        return '2nd-' + str(minute)

    
df['time'] = df.apply(lambda x: create_xticks(x['period'], x['minute']), axis=1)
df.head()
```

```python
shots_tidy = df.groupby(['team', 'period', 'minute', 'time', 'shot_outcome']).sum().groupby(level=0).cumsum().reset_index()
shots_tidy.head()
```

```python
shots_tidy = shots_tidy.sort_values(['period', 'minute']).reset_index()
shots_tidy
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

sns.lineplot('time', 'shot_statsbomb_xg', hue='team', data=shots_tidy, ax=ax)

ax.title.set_text('xG Belgium vs Japan - FIFA World Cup 2018')

plt.xticks(rotation=45)
plt.show()
```

```python
#split away team
ateam = shots_tidy[shots_tidy['team']=='Japan']
```

```python
#split home team
hteam = shots_tidy[shots_tidy['team']=='Belgium']
```

```python
fig, ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

#set up our base layer
mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'

ax.grid(ls='dotted',lw=.5,color='lightgrey',axis='y',zorder=1)
spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)
        
plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Minute',fontname='Andale Mono',color='white',fontsize=16)
plt.ylabel('xPected Goals',fontname='Andale Mono',color='white',fontsize=16)

#plot the step graphs
ax.step(x=ateam['minute'],y=ateam['shot_statsbomb_xg'],color='#d3d3d3',label=ateam['team'],linewidth=5,where='post')
ax.step(x=hteam['minute'],y=hteam['shot_statsbomb_xg'],color='#fd3607',label=hteam['team'],linewidth=5,where='post')
```
