```python
df['pass_outcome'].fillna("Completed", inplace=True)
df_pass = df['pass_outcome']
df_pass
```

```python
#decide what columns to include before analyzing the data 
df = df[['location', 'pass_end_location', 'team', 'player', 'position', 'timestamp', 'pass_outcome']].dropna()
argentina_df = df.loc[df['team']=='Argentina'].reset_index(drop=True)
argentina_df
```

```python
enzo_fc = argentina_df.loc[argentina_df['player']=='Enzo Nicolás Pérez'].reset_index(drop=True)
enzo_fc
```

```python
#split location and pass_end_location to form of data x and y for each person
start_df = pd.DataFrame(enzo_fc['location'].to_list(), columns=['start_x', 'start_y']).reset_index(drop=True)
enzo_all = enzo_fc.join(start_df)
end_df = pd.DataFrame(enzo_fc['pass_end_location'].to_list(), columns=['end_x', 'end_y']).reset_index(drop=True)
enzo_all = enzo_all.join(end_df)
enzo_all
```

```python
#make progressive pass analysis
#Import mplsoccer for convenience for pitch draw
from mplsoccer import Pitch
```

```python
#rescaling for pitch statsbomb
#enzo_all.start_x = enzo_all.start_x*1.2
#enzo_all.start_y = enzo_all.start_y*.8
#enzo_all.end_x = enzo_all.end_x*1.2
#enzo_all.end_y = enzo_all.end_y*1.2
```

```python
#making new columns for proggressive analysis
enzo_all['beginning'] = np.sqrt(np.square(120-enzo_all['start_x']) + np.square(40 - enzo_all['start_y']))
enzo_all['end'] = np.sqrt(np.square(120-enzo_all['end_x']) + np.square(40 - enzo_all['end_y']))

enzo_all['progressive'] = [(enzo_all['end'][x]) / (enzo_all['beginning'][x]) < .75 for x in range(len(enzo_all.beginning))]
enzo_all
```

```python
#loc all pass_progressive True
pass_progressive = enzo_all.loc[(enzo_all['pass_outcome']=='Completed') & (enzo_all['progressive']==True)]
pass_progressive
```

```python
#draw the pitch
pitch = Pitch(pitch_type = 'statsbomb')
fig, ax = pitch.draw()

pitch.lines(enzo_all.start_x, enzo_all.start_y, enzo_all.end_x, enzo_all.end_y, comet = True, ax=ax)
```
